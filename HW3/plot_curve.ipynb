```python
import matplotlib.pyplot as plt
import numpy as np
import json 
```

```python
log_file = 'OUTPUTS/300steps/trainer_state.json'
with open(log_file,'r') as file:
    logs = json.load(file)
```

```python
table = None
for log in logs['log_history']:
    if 'eval_loss' not in log.keys():
        continue
    point = np.array((log['step'],log['eval_loss']))
    if table is None:
        table = point
    else:
        table = np.vstack((table,point))
```

```python
plt.plot(table[:,0], table[:,1])
plt.title("Loss curve of public test set")
plt.xlabel("step")
plt.ylabel("loss")
plt.savefig("Loss.png")
plt.show()
```
